## Missing Data Management

Missing data in this data set is giving by blanks, rather than NAs or 0s, so the strategy will be to find them depending on the type of variable of each column.

```r

results <- lapply(names(data), function(col_name) {
  col <- data[[col_name]]
  if (is.character(col)) {
    count <- sum(col == "" | is.na(col))
    percentage <- (count / length(col)) * 100
  } else if (is.integer(col)) {
    count <- sum(is.na(col))
    percentage <- (count / length(col)) * 100
  } else {
    count <- 0
    percentage <- 0
  }
  data.frame(Column = col_name, Count = count, Percentage = round(percentage, 2))
})
result_table <- do.call(rbind, results)
print(result_table)
```

Lot Frontage is related with Lot Config, so we can replace with 0 properties that are inside lots.

```r
install.packages("ggplot2")
library(ggplot2)
if (!require('dplyr')) install.packages("dplyr")
library('dplyr')
data2 <- data
data2$Lot.Frontage[is.na(data$Lot.Frontage) & data$Lot.Config == "Inside"] <- 0
count(data2, is.na(data2$Lot.Frontage) )
```

For the remaining 216 missing values, we will calculate the average ratio between lot.Frontage and Lot.Area by Lot.Config and replace them.

```r
mean_ratios <- data2 %>%
  group_by(Lot.Config) %>%
  summarize(mean_ratio = mean(Lot.Frontage / Lot.Area, na.rm = TRUE))
data2 <- merge(data2, mean_ratios, by = "Lot.Config", all.x = TRUE)
data2$Lot.Frontage <- ifelse(is.na(data2$Lot.Frontage), data2$mean_ratio * data2$Lot.Area, data2$Lot.Frontage)
data2$mean_ratio <- NULL
count(data2, is.na(data2$Lot.Frontage) )
```

The information is only mostly missing regarding Bsmt, so we explore further. We decide to replace this row's missing values with the mode grouping houses in the same Neighborhood, House Style, and built in the same decade.

```r

data_subset <- data2 %>%
  filter(Year.Built >= 1940 & Year.Built <= 1949)


cols_to_impute <- c("Bsmt.Qual", "Bsmt.Cond", "Bsmt.Exposure", "BsmtFin.Type.1",
                    "BsmtFin.SF.1", "BsmtFin.Type.2", "BsmtFin.SF.2", "Bsmt.Unf.SF",
                    "Total.Bsmt.SF", "Bsmt.Full.Bath", "Bsmt.Half.Bath")


for (col in cols_to_impute) {

  mode_values <- data_subset %>%
    group_by(Neighborhood, House.Style) %>%
    summarise(mode_val = names(which.max(table(get(col)))))


  data2 <- left_join(data2, mode_values, by = c("Neighborhood", "House.Style"))


  data2[data2$Order == 1342, col] <- data2[data2$Order == 1342, "mode_val"]


  data2$mode_val <- NULL
}
data2[data2$Order == 1342, columns_to_check   ]
```

## Before modeling we will create some alternative transformations of our longer categorical values such as Neigborhood, Yr Built and Yr Remodeling
###In particular:
We transformed three variables to improve predictive power for SalePrice:

**Year.Built** was replaced with Antiquity (calculated as Yr.Sold - Year.Built), as the age of a house often influences its value.

**Year.Remod.Add** was transformed into a dummy variable (Remod_5yrs) indicating if remodeling occurred within 5 years, capturing the perceived freshness of updates.

```r

run_ols <- function(data, regressors, response = "SalePrice") {
  formula <- as.formula(paste(response, "~", paste(regressors, collapse = " + ")))
  model <- lm(formula, data = data)


  r_squared <- summary(model)$r.squared
  mse <- mean(residuals(model)^2)

  list(
    model = model,
    model_summary = summary(model),
    r_squared = r_squared,
    mse = mse
  )
}


cat("Model 1: Comprehensive\n")
model_1_results <- run_ols(data5_dummies, regressors_1)
print(model_1_results$model_summary)
cat(sprintf("R^2: %.4f\n", model_1_results$r_squared))
cat(sprintf("MSE: %.4f\n", model_1_results$mse))

cat("\nModel 2: Moderate\n")
model_2_results <- run_ols(data5_dummies, regressors_2)
print(model_2_results$model_summary)
cat(sprintf("R^2: %.4f\n", model_2_results$r_squared))
cat(sprintf("MSE: %.4f\n", model_2_results$mse))

cat("\nModel 3: Simple\n")
model_3_results <- run_ols(data5_dummies, regressors_3)
print(model_3_results$model_summary)
cat(sprintf("R^2: %.4f\n", model_3_results$r_squared))
cat(sprintf("MSE: %.4f\n", model_3_results$mse))



```

```r
# Function to run Lasso regression multiple times on a given dataset
run_lasso_multiple_times <- function(x_data, y_data, num_runs = 100) {
  coefficients_list <- list()
  lambda_list <- numeric(num_runs)
  mse_list <- numeric(num_runs)
  r2_list <- numeric(num_runs)

  for (i in 1:num_runs) {
    # Split data into training and testing sets
    lasso_train_indices <- sample(1:nrow(x_data), size = 0.8 * nrow(x_data))
    x_train <- x_data[lasso_train_indices, ]
    y_train <- y_data[lasso_train_indices]
    x_test <- x_data[-lasso_train_indices, ]
    y_test <- y_data[-lasso_train_indices]

    # Cross-validated Lasso
    cv_lasso <- cv.glmnet(x_train, y_train, alpha = 1)
    best_lambda <- cv_lasso$lambda.min
    lambda_list[i] <- best_lambda

    # Refit the model with the best lambda
    final_lasso <- glmnet(x_train, y_train, alpha = 1, lambda = best_lambda)

    # Predict and calculate MSE and R^2
    predictions <- as.vector(predict(final_lasso, s = best_lambda, newx = x_test))
    mse_list[i] <- mean((y_test - predictions)^2)
    sst <- sum((y_test - mean(y_test))^2)
    sse <- sum((y_test - predictions)^2)
    r2_list[i] <- 1 - (sse / sst)

    # Extract non-zero coefficients
    coefficients <- coef(final_lasso, s = best_lambda)
    coefficients_df <- as.data.frame(as.matrix(coefficients))
    colnames(coefficients_df) <- c("Coefficient")
    coefficients_df$Feature <- rownames(coefficients_df)
    non_zero_coefficients <- coefficients_df[coefficients_df$Coefficient != 0, "Feature"]

    # Store non-zero coefficients from this run
    coefficients_list[[i]] <- non_zero_coefficients
  }

  # Flatten the list of coefficients and count occurrences
  all_coefficients <- unlist(coefficients_list)
  coefficients_count <- table(all_coefficients)

  # Convert to a sorted data frame for better visualization
  coefficients_count_df <- as.data.frame(coefficients_count)
  colnames(coefficients_count_df) <- c("Feature", "Count")
  coefficients_count_df <- coefficients_count_df[order(-coefficients_count_df$Count), ]

  # Aggregate lambda, MSE, and R^2 statistics
  lambda_avg <- mean(lambda_list)
  mse_avg <- mean(mse_list)
  r2_avg <- mean(r2_list)

  list(coefficients_count_df = coefficients_count_df,
       lambda_avg = lambda_avg,
       mse_avg = mse_avg,
       r2_avg = r2_avg)
}

# Run Lasso for each dataset
cat("Running Lasso Regression on Model 1 (Comprehensive)...\n")
results_model1 <- run_lasso_multiple_times(x_lasso1, y_lasso1)

cat("Running Lasso Regression on Model 2 (Moderate)...\n")
results_model2 <- run_lasso_multiple_times(x_lasso2, y_lasso2)

cat("Running Lasso Regression on Model 3 (Simple)...\n")
results_model3 <- run_lasso_multiple_times(x_lasso3, y_lasso3)

# Print the results for each model
cat("Results for Model 1 (Comprehensive):\n")
print(results_model1$coefficients_count_df)
cat(sprintf("Average Optimal Lambda: %.4f\n", results_model1$lambda_avg))
cat(sprintf("Average MSE: %.4f\n", results_model1$mse_avg))
cat(sprintf("Average R^2: %.4f\n", results_model1$r2_avg))

cat("\nResults for Model 2 (Moderate):\n")
print(results_model2$coefficients_count_df)
cat(sprintf("Average Optimal Lambda: %.4f\n", results_model2$lambda_avg))
cat(sprintf("Average MSE: %.4f\n", results_model2$mse_avg))
cat(sprintf("Average R^2: %.4f\n", results_model2$r2_avg))

cat("\nResults for Model 3 (Simple):\n")
print(results_model3$coefficients_count_df)
cat(sprintf("Average Optimal Lambda: %.4f\n", results_model3$lambda_avg))
cat(sprintf("Average MSE: %.4f\n", results_model3$mse_avg))
cat(sprintf("Average R^2: %.4f\n", results_model3$r2_avg))
```

# Predictive power for above/below the mean sale price

In this analysis, we evaluated the predictive power of logistic regression for classifying properties as above or below the mean sales price, using three different regressor sets (comprehensive, moderate, and simple). The rationale was to assess how well models capture complex relationships while balancing simplicity and interpretability.

```r
# Install necessary libraries if not installed
if (!require(pROC)) install.packages("pROC")

library(dplyr)
library(pROC)

set.seed(123)

# Prepare the target variable (above/below mean SalePrice)
mean_price <- mean(data5_dummies$SalePrice, na.rm = TRUE)
data5_dummies <- data5_dummies %>%
  mutate(SalePrice_Binary = ifelse(SalePrice > mean_price, 1, 0))

# Split data into training and test sets
train_index <- sample(1:nrow(data5_dummies), size = 0.8 * nrow(data5_dummies))
train_data <- data5_dummies[train_index, ]
test_data <- data5_dummies[-train_index, ]

# Function to run a logistic regression model and calculate metrics
run_logit <- function(regressors, train_data, test_data) {
  formula <- as.formula(paste("SalePrice_Binary ~", paste(regressors, collapse = " + ")))
  logit_model <- glm(formula, data = train_data, family = "binomial")

  predictions_prob <- predict(logit_model, newdata = test_data, type = "response")
  predictions_class <- ifelse(predictions_prob > 0.5, 1, 0)

  r_squared <- 1 - (logit_model$deviance / logit_model$null.deviance)
  mse <- mean((test_data$SalePrice_Binary - predictions_prob)^2)

  roc_obj <- roc(test_data$SalePrice_Binary, predictions_prob)
  auc <- auc(roc_obj)

  list(r_squared = r_squared, mse = mse, auc = auc, roc_obj = roc_obj)
}

# Run models for the three sets of regressors
logit_model_1 <- run_logit(regressors_1, train_data, test_data)
logit_model_2 <- run_logit(regressors_2, train_data, test_data)
logit_model_3 <- run_logit(regressors_3, train_data, test_data)

# Print results
cat("Logit Model Results:\n")
cat(sprintf("Model 1 (Comprehensive) R^2: %.4f, MSE: %.4f, AUC: %.4f\n", logit_model_1$r_squared, logit_model_1$mse, logit_model_1$auc))
cat(sprintf("Model 2 (Moderate) R^2: %.4f, MSE: %.4f, AUC: %.4f\n", logit_model_2$r_squared, logit_model_2$mse, logit_model_2$auc))
cat(sprintf("Model 3 (Simple) R^2: %.4f, MSE: %.4f, AUC: %.4f\n", logit_model_3$r_squared, logit_model_3$mse, logit_model_3$auc))

# Plot ROC curves
plot(logit_model_1$roc_obj, col = "blue", lwd = 2, main = "ROC Curves: Above/Below Mean SalePrice")
lines(logit_model_2$roc_obj, col = "green", lwd = 2)
lines(logit_model_3$roc_obj, col = "red", lwd = 2)
legend("bottomright", legend = c("Model 1 (Comprehensive)", "Model 2 (Moderate)", "Model 3 (Simple)"),
       col = c("blue", "green", "red"), lwd = 2)
```

```r
# Function to plot confusion matrix
plot_confusion_matrix_box <- function(predictions_class, actual, model_name) {
  cm <- table(Predicted = predictions_class, Actual = actual)
  total <- sum(cm) # Total observations
  cm_percent <- round((cm / total) * 100, 1) # Percentages based on total observations

  cm_df <- as.data.frame(as.table(cm)) # Convert counts to data frame
  cm_df$Percentage <- as.vector(cm_percent) # Add percentage column

  ggplot(cm_df, aes(x = Actual, y = Predicted, fill = Percentage)) +
    geom_tile(color = "white") +
    geom_text(aes(label = paste0(Percentage, "%\n(", Freq, ")")), size = 5) +
    scale_fill_gradient(low = "lightblue", high = "blue") +
    theme_minimal() +
    labs(
      title = paste("Confusion Matrix (%):", model_name),
      x = "Actual Class",
      y = "Predicted Class",
      fill = "Percentage"
    ) +
    theme(
      plot.title = element_text(size = 14, face = "bold"),
      axis.title = element_text(size = 12),
      axis.text = element_text(size = 10)
    )
}

# Predictions for each model
logit_predictions_1 <- ifelse(predict(glm(as.formula(paste("SalePrice_Binary ~", paste(regressors_1, collapse = " + "))),
                                         data = train_data, family = "binomial"), newdata = test_data, type = "response") > 0.5, 1, 0)

logit_predictions_2 <- ifelse(predict(glm(as.formula(paste("SalePrice_Binary ~", paste(regressors_2, collapse = " + "))),
                                         data = train_data, family = "binomial"), newdata = test_data, type = "response") > 0.5, 1, 0)

logit_predictions_3 <- ifelse(predict(glm(as.formula(paste("SalePrice_Binary ~", paste(regressors_3, collapse = " + "))),
                                         data = train_data, family = "binomial"), newdata = test_data, type = "response") > 0.5, 1, 0)

# Plot confusion matrices for each model
print(plot_confusion_matrix_box(logit_predictions_1, test_data$SalePrice_Binary, "Model 1 (Comprehensive)"))
print(plot_confusion_matrix_box(logit_predictions_2, test_data$SalePrice_Binary, "Model 2 (Moderate)"))
print(plot_confusion_matrix_box(logit_predictions_3, test_data$SalePrice_Binary, "Model 3 (Simple)"))


```


The confusion matrices reveal that while logistic regression performs reasonably well, its predictive power is limited, particularly for the "above mean" class (low percentages of correct predictions). Gradient Boosting (GB) and OLS are superior as they better capture the underlying relationships, achieving higher R2 and lower MSE in prior evaluations. These results suggest that logistic regression may struggle with nuanced, continuous target variables compared to GB and OLS.
